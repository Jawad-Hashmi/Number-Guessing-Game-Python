```python
import random

# random number generation for guessing
number = random.randint(1,100)

# Number guessed by user
guess = int(input('Please Enter Your Number'))
counter = 1
# Applying loop for giving more than one attempts to user for guessing
while guess != number:
    # Now validating the logic for the number guessed by user
    if guess < number:
            print ("Wrong! Guess Higher")
    else:
            print ('Wrong! Guess Lower')
        
    guess = int(input("Please Enter Your Number"))
    counter += 1

# if the upper condition are wong mean the guess number is corerct so 
else:
    print ("Correct The Number is ", guess)
    print('Attempts', counter)


    

```
